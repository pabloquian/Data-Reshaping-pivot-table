#### What's the mean salary by department?
Hint : There are 2 ways to do it - one way is with groupby and the other way is with pivot tables. Do it both ways. 


```python
# Method 1 groupby
display(employee.groupby(['Department']).agg({'Salary':pd.Series.mean}))
```

```python
# Method 2 pivot_table
employee.pivot_table(index=['Department'], values = ['Salary'],aggfunc=['mean'])
```

#### We want now to group the data by more than one column. Use Pandas pivot table function to select the mean salary by department and title.

```python
employee.pivot_table(index= ['Department','Title'],values = ['Salary'],aggfunc=['mean'])
```

#### We want to display the number of years of work experience and the salary for each Title. Add Years as a column in your pivot table.
*Hint: Use Pandas doc(see References in README.md) to explore the pivot_table() arguments*

```python
employee.pivot_table(index= ['Title'],values = ['Salary', 'Years'],aggfunc=['mean','sum'])
```

#### We want to know the information about the mean salary and number of employees in each department and for each job title.

```python
display(employee.groupby(['Title']).agg({'Salary':pd.Series.mean, 'Name':'count'}))
```

#### We want to know for each department and for each job title the Salary's median, minimum, maximum and standard deviation. Comment your results.

```python
employee.pivot_table(index= ['Department','Title'],values = ['Salary'],aggfunc=['mean', 'min', 'max', 'std', 'median'])
```

```python
# I see that  for some jobs in some departments, there is just one salary
# Those are VP at the HR and IT departments (which also are the best payed), and associate in the IT
# and Sales departments.
# Also, we see that the std is low (in the jobs that have std>0). This is because I am working with
# a dataset with only 10 rows, and there is just one or two people that have the same job in the same department.
```

#### Based on your comments, fill the missing values with an appropriate value.

```python
# There are no missing values. But ,if we talk about the table above, those NaN values represent that
# there is no deviation as there is only one employee with that job in that department.
# That's because the formula of std needs more than one observation to get a 'normal' result.
# The formula of std divides by n-1 and 1-1=0, and something divided by 0 is infinite.
```

```python
employee.pivot_table(index= ['Department','Title'],values = ['Salary'],aggfunc=['mean', 'min', 'max', 'std', 'median']).fillna(0)
```

### Bonus

#### We now want to compute the mean salary after removing the maxima for each Department. Create the appropriate pivot table.

Hint: Write a custom function.

```python
def remv_max_then_mean(x):
    return x.drop(x.idxmax()).mean()
```

```python
employee.pivot_table(index= ['Department'], values = ['Salary'], aggfunc = remv_max_then_mean)
```
